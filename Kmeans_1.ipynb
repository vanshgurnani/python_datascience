**KMeans-1**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline 
```

```python
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data[1][5]
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
mean=KMeans(n_clusters=4)
```

```python
mean.fit(data[0])
```

```python
mean.cluster_centers_
```

```python
mean.labels_
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('KMeans')
ax1.scatter(data[0][:,0],data[0][:,1],c=mean.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

**KMeans-2**

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
%matplotlib inline 
```

```python
df=pd.read_csv('/content/College_Data',index_col=0)
df.head(2)
```

```python
sns.set_style('whitegrid')
sns.lmplot('Room.Board','Grad.Rate',data=df, hue='Private',
           palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

```python
sns.set_style('whitegrid')
sns.lmplot('Outstate','F.Undergrad',data=df, hue='Private',
           palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

```python
df[df['Grad.Rate']>100]
```

```python
df['Grad.Rate']['Cazenovia College']=100
```

```python
df[df['Grad.Rate']>100]
```

```python
from sklearn.cluster import KMeans
```

```python
mean=KMeans(n_clusters=2)
```

```python
mean.fit(df.drop('Private',axis=1))
```

```python
mean.cluster_centers_
```

```python
def school(cluster):
  if cluster=='Yes':
    return 1
  else :
    return 0
```

```python
df['Cluster']=df['Private'].apply(school)
```

```python
df.head(2)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(df['Cluster'],mean.labels_))
print('\n')
print(classification_report(df['Cluster'],mean.labels_))
```
