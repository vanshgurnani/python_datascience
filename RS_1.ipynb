```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 
```

```python
col_name=['user_id','item_id','rating','timestamp']
df=pd.read_csv('u.data',sep='\t',names=col_name)
```

```python
df.head(2)
```

```python
movie_titles=pd.read_csv('Movie_Id_Titles')
```

```python
df=pd.merge(df,movie_titles,on='item_id')
```

```python
df.head(2)
```

```python
sns.set_style('white')
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rating=pd.DataFrame(df.groupby('title')['rating'].mean())
rating.head(2)
```

```python
rating['number']=pd.DataFrame(df.groupby('title')['rating'].count())
rating.head(2)
```

```python
moviemat=df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
moviemat.head(2)
```

```python
rating['number'].sort_values(ascending=False).head(10)
```

```python
star_rate=moviemat['Star Wars (1977)']
liar_rate=moviemat['Liar Liar (1997)']
star_rate.head()
```

```python
similar_star=moviemat.corrwith(star_rate)
```

```python
similar_liar=moviemat.corrwith(liar_rate)
```

```python
corr_star=pd.DataFrame(similar_star,columns=['correlation'])
corr_star.dropna(inplace=True)
corr_star.sort_values('correlation',ascending=False).head(10)
```

```python
corr_star=corr_star.join(rating['number'])
```

```python
corr_star.head(4)
```

```python
corr_star[corr_star['number']>100].sort_values('correlation',ascending=False).head(4)
```

```python
corr_liar=pd.DataFrame(similar_liar,columns=['correlation'])
corr_liar.dropna(inplace=True)
corr_liar.sort_values('correlation',ascending=False).head(10)
```

```python
corr_liar=corr_liar.join(rating['number'])
```

```python
corr_liar[corr_liar['number']>100].sort_values('correlation',ascending=False).head(4)
```
